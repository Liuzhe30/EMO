```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.metrics import roc_curve, auc
```

```python
import scipy.stats as sci
def pearson(x,y):
    r,p = sci.pearsonr(x,y)
```

```python
sns.set_style('ticks')
target = np.load('../model/pred_results/small/predict.npy', allow_pickle=True)
slope = np.load('../model/test_slope/small_slope.npy', allow_pickle=True)
pred_score = target[:,1]

re_x = slope.reshape(slope.shape[0],1)
re_y = pred_score.reshape(slope.shape[0],1)

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(re_x,re_y)
inter = reg.intercept_
coef = reg.coef_
r2 = reg.score(re_x,re_y)

plt.title('KDE between slope and prediction score (Small Model)')
#sns.lmplot(x_estimator=np.mean,x=slope,y=pred_score)
x = np.linspace(-2.1,2.3,4)
y = coef[0][0]*x + inter[0]
plt.plot(x,y,color='#000',linestyle='--')
ax = sns.kdeplot(x=slope,y=pred_score,color='#AFEEEE',shade=True,cbar=True)
corr, p = sci.pearsonr(slope, pred_score)
ax.annotate(f' R square = {r2:.3f} \n Corr = {corr:.3f} \n p < 0.001', (0.8, 0.05), xycoords='axes fraction', ha='center')
plt.ylabel("Prediction score")
plt.xlabel("Slope of eQTL")
plt.savefig('images/KDE/small.png',dpi=300)
plt.show()
```
