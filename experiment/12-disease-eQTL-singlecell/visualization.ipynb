```python
import matplotlib.pyplot as plt
import numpy as np

values = []  
labels = ['Dendritic_Cell','B_Naive','CD4_memory','CD4_naive','Natural_killer_Cell','CD8_memory']
for cell in labels:
    values.append(np.load('data/rs1465697_large_predict_' + cell + '.npy')[0][0])

# Define custom colors for each bar
colors = ['#E57961','#E39844', '#949BBF', '#A3A0C3', '#7FBA80', '#79ABC9']
new_labels = ['Dendritic Cell','B Naive','CD4+ T memory','CD4+ T naive','Natural killer Cell','CD8 memory']

# Create a bar plot
fig, ax = plt.subplots(figsize=(5, 3))
bars = ax.bar(labels, values, color=colors, edgecolor='black',alpha=0.6)

# Add black borders and values on top of the bars
for bar, value in zip(bars, values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{value:.2f}', ha='center', va='bottom')

# Add labels and title
ax.set_ylabel('Predicted slope')
ax.set_title('Pattern prediction of rs1465697-T to CCDC155')
ax.set_xticklabels(new_labels, rotation=25, ha="right")
plt.ylim((-0.5, 0.68))

# Show the plot
plt.tight_layout()
plt.savefig('barplot.png',dpi=300, bbox_inches = 'tight')
plt.show()
```
