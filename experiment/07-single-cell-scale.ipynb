#### single-cell scale

```python
import numpy as np
import pandas as pd
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import confusion_matrix, precision_score, accuracy_score,recall_score, f1_score,roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load datasets
y_true = np.load('../model/pred_results_sc/large_B_Naive_label.npy')
y_pred = np.load('../model/pred_results_sc/large_B_Naive_predict.npy')

y_score = y_pred.argmax(axis=1)       
test_label = y_true.argmax(axis=1)                         
y_score_pro = y_pred # (212, 2)
y_one_hot = y_true
y_score_one_hot = to_categorical(y_score)

obj1 = confusion_matrix(test_label, y_score)
print('confusion_matrix\n', obj1)
print('accuracy:{}'.format(accuracy_score(test_label, y_score)))
print('precision:{}'.format(precision_score(test_label, y_score)))
print('recall:{}'.format(recall_score(test_label, y_score)))
print('f1-score:{}'.format(f1_score(test_label, y_score)))

# update confusion matrix
sum_all1 = obj1[0][0] + obj1[0][1]
sum_all2 = obj1[1][0] + obj1[1][1]
new_obj = np.array(
    [
        [float(obj1[0][0])/sum_all1, float(obj1[0][1])/sum_all1],
        [float(obj1[1][0])/sum_all2, float(obj1[1][1])/sum_all2]
    ]
)

sns.heatmap(new_obj, fmt='.2f', cmap='Blues', annot=True)
plt.ylabel("true label")
plt.xlabel("predict label")
plt.savefig('images/singlecell/cm_B_Naive.png',dpi=300)
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
fpr_xgb, tpr_xgb, thresholds = roc_curve(y_one_hot.ravel(),y_score_pro.ravel())   
auc_xgb = auc(fpr_xgb, tpr_xgb)
print("AUC : ", auc_xgb)

plt.plot(fpr_xgb, tpr_xgb, linewidth = 2,label='AUC=%.3f' % auc_xgb)
plt.plot([0,1],[0,1], 'k--')
plt.axis([0,1.1,0,1.1])
plt.xlabel('False Postivie Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.savefig('images/singlecell/ROC_B_Naive.png',dpi=300)
plt.show()
```

#### test representation

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn import datasets
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA

import warnings
warnings.filterwarnings('ignore')

def plot_embedding(data, label, title):
    x_min, x_max = np.min(data, 0), np.max(data, 0)
    data = (data - x_min) / (x_max - x_min)

    fig = plt.figure()
    ax = plt.subplot(111)
    for i in range(data.shape[0]):
        plt.text(data[i, 0], data[i, 1], str(label[i]),
                 color=plt.cm.Set1(label[i] / 10.),
                 fontdict={'weight': 'bold', 'size': 9})
    plt.xticks([])
    plt.yticks([])
    plt.title(title)
    return fig
```

```python
# load datasets
input1 = np.load("../model/middle_output_sc/B_Naive_input.npy",allow_pickle=True) 
output1 = np.load("../model/middle_output_sc/B_Naive_output.npy",allow_pickle=True) 
input2 = np.load("../model/middle_output_sc/CD4_memory_input.npy",allow_pickle=True) 
output2 = np.load("../model/middle_output_sc/CD4_memory_output.npy",allow_pickle=True) 
input3 = np.load("../model/middle_output_sc/CD4_naive_input.npy",allow_pickle=True) 
output3 = np.load("../model/middle_output_sc/CD4_naive_output.npy",allow_pickle=True) 
input4 = np.load("../model/middle_output_sc/CD8_memory_input.npy",allow_pickle=True) 
output4 = np.load("../model/middle_output_sc/CD8_memory_output.npy",allow_pickle=True) 
input5 = np.load("../model/middle_output_sc/Dendritic_Cell_input.npy",allow_pickle=True) 
output5 = np.load("../model/middle_output_sc/Dendritic_Cell_output.npy",allow_pickle=True) 
input6 = np.load("../model/middle_output_sc/Natural_killer_Cell_input.npy",allow_pickle=True) 
output6 = np.load("../model/middle_output_sc/Natural_killer_Cell_output.npy",allow_pickle=True) 
```

```python
label_merge = []
for i in range(input1.shape[0]):
    label_merge.append(0)
for i in range(input2.shape[0]):
    label_merge.append(1)
for i in range(input3.shape[0]):
    label_merge.append(2)
for i in range(input4.shape[0]):
    label_merge.append(3)
for i in range(input5.shape[0]):
    label_merge.append(4)
for i in range(input6.shape[0]):
    label_merge.append(5)
label = np.array(label_merge)
print(label.shape)
```

```python
input = np.concatenate([input1,input2,input3,input4,input5,input6],axis=0)
output = np.concatenate([output1,output2,output3,output4,output5,output6],axis=0)
```

```python
stack = input

fig = plt.figure()
tsne = TSNE(n_components=2, init='pca', random_state=0)
print(stack.shape)
result = tsne.fit_transform(stack)
print(result.shape)
#fig = plot_embedding(result, label,'t-SNE embedding of the digits')
x_min, x_max = np.min(result, 0), np.max(result, 0)
result = (result - x_min) / (x_max - x_min)

color = ["#B278A7","#7CB0DF","#8FCA88","#F3EE58","#FF8C00","#800000"]

ax = plt.subplots(figsize=(5,5))
for i in range(result.shape[0]):
    if(label[i] == 0):
        s1 = plt.scatter(result[i, 0], result[i, 1],s=10,color=color[label[i]])
for i in range(result.shape[0]):
    if(label[i] == 1):
        s2 = plt.scatter(result[i, 0], result[i, 1],s=10,color=color[label[i]])
for i in range(result.shape[0]):
    if(label[i] == 2):
        s3 = plt.scatter(result[i, 0], result[i, 1],s=10,color=color[label[i]])
for i in range(result.shape[0]):
    if(label[i] == 3):
        s4 = plt.scatter(result[i, 0], result[i, 1],s=10,color=color[label[i]])
for i in range(result.shape[0]):
    if(label[i] == 4):
        s5 = plt.scatter(result[i, 0], result[i, 1],s=10,color=color[label[i]])
for i in range(result.shape[0]):
    if(label[i] == 5):
        s6 = plt.scatter(result[i, 0], result[i, 1],s=10,color=color[label[i]])
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title("Representation of input of cell types (Large Model)")
plt.legend((s1,s2,s3,s4,s5,s6),('B_Naive','CD4_memory','CD4_naive','CD8_memory','Dendritic_Cell','Natural_killer_Cell') ,loc = 'best')
plt.savefig('images/tsne/sc-input.png',dpi=300)
plt.show()
```
