#### metabrain evaluation

####  1 finetune

```python
import numpy as np
import pandas as pd
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import confusion_matrix, precision_score, accuracy_score,recall_score, f1_score,roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load datasets
y_true = np.load('../model/pred_results_metabrain/middle_hippocampus_label.npy')
y_pred = np.load('../model/pred_results_metabrain/middle_hippocampus_predict.npy')

y_score = y_pred.argmax(axis=1)       
test_label = y_true.argmax(axis=1)                         
y_score_pro = y_pred # (212, 2)
y_one_hot = y_true
y_score_one_hot = to_categorical(y_score)

obj1 = confusion_matrix(test_label, y_score)
print('confusion_matrix\n', obj1)
print('accuracy:{}'.format(accuracy_score(test_label, y_score)))
print('precision:{}'.format(precision_score(test_label, y_score)))
print('recall:{}'.format(recall_score(test_label, y_score)))
print('f1-score:{}'.format(f1_score(test_label, y_score)))

# update confusion matrix
sum_all1 = obj1[0][0] + obj1[0][1]
sum_all2 = obj1[1][0] + obj1[1][1]
new_obj = np.array(
    [
        [float(obj1[0][0])/sum_all1, float(obj1[0][1])/sum_all1],
        [float(obj1[1][0])/sum_all2, float(obj1[1][1])/sum_all2]
    ]
)
ax = plt.subplots(figsize=(3,3))
plt.title('Medium Model, Hippocampus\nFine-tune (20 test samples)')
sns.heatmap(new_obj, fmt='.2f', cmap='bone_r', annot=True)
plt.ylabel("true label")
plt.xlabel("predict label")
plt.savefig('images/metabrain/cm_middle_hippocampus.png',dpi=300, bbox_inches = 'tight')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
fpr_xgb, tpr_xgb, thresholds = roc_curve(y_one_hot.ravel(),y_score_pro.ravel())   
auc_xgb = auc(fpr_xgb, tpr_xgb)
print("AUC : ", auc_xgb)
ax = plt.subplots(figsize=(3,3))
plt.title('Medium Model, Hippocampus\nFine-tune')
plt.plot(fpr_xgb, tpr_xgb, linewidth = 2,label='AUC=%.3f' % auc_xgb,color='#778899')
plt.plot([0,1],[0,1], 'k--')
plt.axis([0,1.1,0,1.1])
plt.xlabel('False Postivie Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.savefig('images/metabrain/ROC_middle_hippocampus.png',dpi=300, bbox_inches = 'tight')
plt.show()
```

```python
# convert and save plot data
import numpy as np
import pandas as pd
import os

output_dir = 'plot_data'
os.makedirs(output_dir, exist_ok=True)

output_file = os.path.join(output_dir, 'Figure6a_6d_plot_data.xlsx')

datasets = [
    ('../model/pred_results_metabrain/middle_hippocampus_label.npy', '../model/pred_results_metabrain/middle_hippocampus_predict.npy', 'middle_hippocampus'),
    ('../model/pred_results_metabrain/middle_spinalcord_label.npy', '../model/pred_results_metabrain/middle_spinalcord_predict.npy', 'middle_spinalcord'),
    ('../model/pred_results_metabrain/large_hippocampus_label.npy', '../model/pred_results_metabrain/large_hippocampus_predict.npy', 'large_hippocampus'),
    ('../model/pred_results_metabrain/large_spinalcord_label.npy', '../model/pred_results_metabrain/large_spinalcord_predict.npy', 'large_spinalcord'),
    ('../model/pred_results_metabrain_e2e/middle_hippocampus_label.npy', '../model/pred_results_metabrain_e2e/middle_hippocampus_predict.npy', 'middle_hippocampus_e2e'),
    ('../model/pred_results_metabrain_e2e/middle_spinalcord_label.npy', '../model/pred_results_metabrain_e2e/middle_spinalcord_predict.npy', 'middle_spinalcord_e2e'),
    ('../model/pred_results_metabrain_e2e/large_hippocampus_label.npy', '../model/pred_results_metabrain_e2e/large_hippocampus_predict.npy', 'large_hippocampus_e2e'),
    ('../model/pred_results_metabrain_e2e/large_spinalcord_label.npy', '../model/pred_results_metabrain_e2e/large_spinalcord_predict.npy', 'large_spinalcord_e2e')
]

with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
    for label_path, predict_path, name in datasets:

        y_true = np.load(label_path)
        y_pred = np.load(predict_path)

        df_true = pd.DataFrame(y_true).fillna(0)
        df_pred = pd.DataFrame(y_pred).fillna(0)
        
        df_true.to_excel(writer, sheet_name=f'{name}_label', index=False)
        df_pred.to_excel(writer, sheet_name=f'{name}_predict', index=False)
```
