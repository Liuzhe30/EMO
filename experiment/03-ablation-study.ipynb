```python
import os
import pandas as pd
import numpy as np
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import confusion_matrix, precision_score, accuracy_score,recall_score, f1_score,roc_auc_score
```

```python
# no ATAC
y_true = np.load('../model/pred_results_ablation/no_ATAC_label.npy')
y_pred = np.load('../model/pred_results_ablation/no_ATAC_predict.npy')

y_score = y_pred.argmax(axis=1)       
test_label = y_true.argmax(axis=1)                         
y_score_pro = y_pred # (212, 2)
y_one_hot = y_true
y_score_one_hot = to_categorical(y_score)

obj1 = confusion_matrix(test_label, y_score)
print('confusion_matrix\n', obj1)
print('accuracy:{}'.format(accuracy_score(test_label, y_score)))
print('precision:{}'.format(precision_score(test_label, y_score)))
print('recall:{}'.format(recall_score(test_label, y_score)))
print('f1-score:{}'.format(f1_score(test_label, y_score)))

from sklearn.metrics import roc_curve, auc
fpr_xgb, tpr_xgb, thresholds = roc_curve(y_one_hot.ravel(),y_score_pro.ravel())   
auc_xgb = auc(fpr_xgb, tpr_xgb)
print("AUC : ", auc_xgb)
```

```python
import numpy as np
import matplotlib.pyplot as plt

results = [{"ACC": 0.762, "Recall": 0.732, "Precision": 0.793, "F1-score": 0.756, "AUC": 0.838},# without ATAC-seq
           {"ACC": 0.589, "Recall": 0.494, "Precision": 0.623, "F1-score": 0.551, "AUC": 0.619},# without DNA sequence
           {"ACC": 0.745, "Recall": 0.701, "Precision": 0.777, "F1-score": 0.737, "AUC": 0.811},# without BiGRU
           {"ACC": 0.724, "Recall": 0.744, "Precision": 0.724, "F1-score": 0.734, "AUC": 0.804},# without attention layers
           {"ACC": 0.807, "Recall": 0.785, "Precision": 0.831, "F1-score": 0.807, "AUC": 0.864}] # EMO-medium
data_length = len(results[0])
angles = np.linspace(0, 2 * np.pi, data_length, endpoint=False)
labels = [key for key in results[0].keys()]
score = [[v for v in result.values()] for result in results]

angles = np.concatenate((angles, [angles[0]]))
labels = np.concatenate((labels, [labels[0]]))
score_1 = np.concatenate((score[0], [score[0][0]]))
score_2 = np.concatenate((score[1], [score[1][0]]))
score_3 = np.concatenate((score[2], [score[2][0]]))
score_4 = np.concatenate((score[3], [score[3][0]]))
score_5 = np.concatenate((score[4], [score[4][0]]))
```

```python
fig = plt.figure(figsize=(3,3))
ax = plt.subplot(polar = True)

plt.plot(angles, score_1, color='#5867AF')
plt.plot(angles, score_5, color='#808080')
plt.fill(angles, score_1, '#5867AF', alpha=0.4, label='_nolegend_')
plt.fill(angles, score_5, '#DCDCDC', alpha=0.6, label='_nolegend_')
ax.set_rlim(0.4, 0.9)

ax.set_theta_zero_location('N') 
ax.set_rlabel_position(0)
ax.set_thetagrids(angles * 180 / np.pi, labels, fontsize=9,rotation=10)
plt.yticks([0.4, 0.5, 0.6, 0.7, 0.8, 0.9], ["0.4", "0.5", "0.6", "0.7", "0.8", "0.9"], color="#696969", size=7)
plt.ylim(0.4, 0.9)
ax.set_title("Without ATAC-seq")
ax.legend(labels=('Without ATAC-seq', 'EMO'), loc = 'center',frameon = True,bbox_to_anchor=(0.5,-0.2))
plt.savefig('images/ablation/atac-seq.png',dpi=300, bbox_inches = 'tight')
plt.show()
```
